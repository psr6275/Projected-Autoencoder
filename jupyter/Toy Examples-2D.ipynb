## Explicit Autoencoder

```python
n_dim = 2
encoding_dim = 1000
input_img = Input(shape=(n_dim,))
encoded = Dense(encoding_dim,activation = 'relu',kernel_initializer = 'glorot_normal')(input_img)

decoder = Dense(n_dim, activation = 'linear',kernel_initializer = 'glorot_normal')
decoded = decoder(encoded)

autoencoder = Model(inputs = input_img, outputs = decoded)
autoencoder.compile(optimizer= 'nadam', loss = 'mse')

encoder = Model(inputs=input_img, outputs=encoded)
```

```python
noise_type = "gaussian"
noise_rep = 1
noise_scale = 0.03
test_size = 0.1
trainX, trainXn = corrupt_toy(xtrain,noise_type = noise_type,rep = noise_rep,scale = noise_scale)
idxs = np.array(range(len(xtrain)))
val_num = int(len(idxs)*test_size)
xtr_o = trainX[idxs[val_num:]]
xtr_n = trainXn[idxs[val_num:]]
xval_o = trainX[idxs[:val_num]]
xval_n = trainXn[idxs[:val_num]]
```

```python
autoencoder.fit(xtr_n,xtr_o, epochs = 300, batch_size = 300, shuffle = True, validation_data = (xval_n,xval_o))
```
